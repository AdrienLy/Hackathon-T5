```python
import requests
import unittest
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
years = range(2000, 2019)

months = ["01","02","03","04","05","06","07","08","09","10","11","12"]

link = "https://www.billboard.com/charts/billboard-200/"

```

```python
def main (date):
    url = f"{link}{date}-01"
    res= requests.get(url)
    if res.status_code == 200:
        html_doc = res.text
        # HTML parsing
        soup_search = BeautifulSoup(html_doc,"html.parser")

        


    # Grabs all information related to the top 200 songs
    containers = soup_search.select('.chart-list-item') # *= means contains

    filename = f'billboard_hot_200_{date}.csv'
    f = open(filename, 'w') # w = write

    headers = 'Song, Artist, Last Week, Peak Position, Weeks on Chart\n'

    f.write(headers)

    #TOP 1 
    song = soup_search.find("div", class_="chart-number-one__title").text
    artist = soup_search.find("div", class_="chart-number-one__artist").text.strip()
    last_week = "-"
    peak_position = "1"
    weeks_in_top = soup_search.find("div", class_="chart-number-one__weeks-on-chart").text.strip()


    f.write('\"' + song + '\",\"' + artist.replace('Featuring', 'Feat.') + '\",' + last_week + ',' + peak_position + ',' + weeks_in_top + '\n')



    chart_position = 1

    # Loops through each container
    for container in containers:

        # Container storing the song name and artist name
        song_container = container.select('.chart-list-item__title-text')
        song = song_container[0].text

        # Grabs the artist name
        try:
            artist = container.select('.chart-list-item__artist > a')
            artist = artist[0].text
        except IndexError:
            artist = container.select('.chart-list-item__artist')
            artist = artist[0].text



        # Grabs the song's position last week
        try:
            last_week = container.select('.chart-list-item__last-week')
            last_week = last_week[0].text
        except IndexError:
            last_week = "-"

        # Grabs the song's peak position
        try:
            peak_position = container.select('.chart-list-item__weeks-at-one')
            peak_position = peak_position[0].text
        except IndexError:
            peak_position = "-"
        # Grabs the song's duration in the hot 200 (in weeks)
        try:
            weeks_in_top = container.select('.chart-list-item__weeks-on-chart')
            weeks_in_top = weeks_in_top[0].text
        except IndexError:
            weeks_in_top = "-"

        # Prints the chart position, song name, artist name, and related stats
        #if print_data:
        #print('\nPosition: #{}'.format(chart_position))
        #print('Song: {}'.format(song))
        #print('Artist: {}'.format(artist))
        #print('Last Week: {}'.format(last_week))
        #print('Peak Position: {}'.format(peak_position))
        #print('Weeks on Chart: {}'.format(weeks_in_top))

        chart_position += 1

        f.write('\"' + song + '\",\"' + artist.replace('Featuring', 'Feat.') + '\",' + last_week + ',' + peak_position + ',' + weeks_in_top + '\n')

    f.close()


```

```python
for year in years:
    for month in months:
        date = f"{year}-{month}"
        main(date)
        
        
```
